```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('all_data.csv')
df.head()
```

Which Countries are represented?

```python
print(df.Country.unique())
```

Which Years are represented?

```python
print(df.Year.unique())
```

Rename Column Header "Life expactancy at birth (years)"

```python
df = df.rename({'Life expectancy at birth (years)': 'LifeExpectancy'}, axis=1)
df
```

```python
df = df.replace(to_replace='United States of America',\
                         value='USA')
```

```python
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, sharey=False, figsize=(15, 4))
axes[0] = sns.boxenplot(ax=axes[0],x="Country", y="LifeExpectancy",color="b", scale="linear", data=df)
axes[0].set_xlabel('Life Expactancy at birth in years')
axes[1] = sns.boxplot(ax=axes[1],x="Country", y="GDP",palette=["m", "g"],data=df)
axes[1].set_xlabel('GDP in billions pet country')

plt.show()
```

```python
sns.set_theme(style="dark")
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(16, 8))
axes[0]=sns.barplot(ax=axes[0],x="LifeExpectancy", y="Country", data=df)
axes[0].set_xlabel("Life expectancy at birth (years)")
axes[1]=sns.barplot(ax=axes[1],x="GDP", y="Country", data=df)
axes[1].set_xlabel("GDP");
```

```python
f, ax = plt.subplots(figsize=(6, 3))
sns.barplot(x="LifeExpectancy", y="Country", data=df)
sns.histplot(x="GDP", y="Country", data=df, bins=35, pthresh=.01, cmap="mako")
#sns.kdeplot(x="Year", y="Country", data=df, levels=5, color="w", linewidths=1)
plt.show()
```

```python
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, sharey=False, figsize=(16, 8))
axes[0] = sns.boxenplot(ax=axes[0],x="Country", y="LifeExpectancy",color="b",data=df)
axes[0] = sns.stripplot(ax=axes[0], x=df.Country, y=df.LifeExpectancy,)
axes[0].set_xlabel('Life Expactancy at birth in years')
axes[1] = sns.boxplot(ax=axes[1],x="Country", y="GDP",palette=["m", "g"],data=df)
axes[1] = sns.stripplot(ax=axes[1], x=df.Country, y=df.GDP)
axes[1].set_xlabel('GDP in billions pet country')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x=df.Year, y=df.GDP, hue=df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("GDP in Trillions of U.S. Dollars");
```

```python
fig, axes = plt.subplots(1, 2, sharey=False, figsize=(15, 7))
axes[0] = sns.lineplot(ax=axes[0],x="Year", y="LifeExpectancy",hue="Country", data=df)
axes[0].set_xlabel('Life Expactancy at birth in years')
axes[0].legend(loc='center left', bbox_to_anchor=(-0.39, 0.5), ncol=1)
axes[1] = sns.lineplot(ax=axes[1],x="Year", y="GDP",hue="Country",data=df)
axes[1].set_xlabel('GDP in billions pet country')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.show()
```
